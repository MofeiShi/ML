```
def scrape_Rank(URL):
  wd = webdriver.Chrome('chromedriver', options = options)
  wd.get(URL)
  soup = bs(wd.page_source,"lxml")
  pullR = soup.findAll('td')

  return pullR[1].string
```

#scrape from wsj

```
def scrape_WSJ(URL):
  wd = webdriver.Chrome('chromedriver', options = options)
  wd.get(URL)
  soup = bs(wd.page_source,"lxml")
  a = soup.findAll('span',class_="data_data")
  print("len(a) is :{}".format(len(a)))
  try:

    out = a[25].text
  except:
    print('An exception has occured')
    out = '-'

  wd.close()
  return out
```

#scrape from cnbc

```
def scrape_CNBC(URL):
  wd = webdriver.Chrome('chromedriver', options = options)
  wd.get(URL)
  soup = bs(wd.page_source,'html.parser')
  try:
    out = soup.find('strong').text
  except:
    out = '-'
    print("An exception has occured.")
  #print(out)
  wd.close()
  return out
```
